#### Recency
- Cluster 3: The lowest recency value is find in this cluster, meaning that customers in this cluster bought recently. Probably, they are **active customers**

- Cluster 0: Have the highest recency value (240 days in average). This indicates that customers in this cluster are **inactive customers**

#### Frequency
- Cluster 3: Apresent the highest frequency value (2631 times in average), indicating that customers in this cluster are **loyal customers**

- Cluster 0: Have the lowes frequency value (26 times in average), indicating that customers in this cluster are **ocasional customers**

#### Monetary
- Cluster 3: Have the highest monetary value (6620 euros in average), indicating that customers in this cluster are **high value customers**

- Cluster 0: Have the lowest monetary value (117 euros in average), suggesting that customers spend less money.


#### Total Orders:
- Cluster 3: Have the highest total orders value (2631 times in average), indicating that customers in this cluster are **big buyers**

- Cluster 0: With fewer orders (26 times in average), reforcing the idea of **inactive customers** or **occasional customers**

#### Average Price:
- Small variations comparing the clusters.

## Conclusion

##### Cluster 0: **Inactive clients** with low frequency and low moneraty spending.
